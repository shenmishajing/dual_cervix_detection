````python
det_path = "/data/luochunhua/cervix/cervix_det/det.pkl"
gt_path = "/data/luochunhua/cervix/cervix_det/gt.pkl"
````

````python
def load_data_from_pkl(pkl_path):
    with open(pkl_path, "rb") as f:
        data = pickle.load(f)
    
    return data 
````

````python
det = load_data_from_pkl(det_path)
gt = load_data_from_pkl(gt_path)
````
